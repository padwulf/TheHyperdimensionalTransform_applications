```julia
function sigmoid(x)
    return 1/(1+exp(-x))
end
"""
function logisticneg(x)
    return -log.(1-sigmoid.(x))
end
function logisticpos(x)
    return -log.(sigmoid.(x))
end
"""

function updateneg(x)
    return x<0 ? 0 : (x)^2
end
function updatepos(x)
    return x>0 ? 0 : (x)^2
end
function logisticneg(x)
    return x + log(1+exp(-x))
end
function logisticpos(x)
    return log(1+exp(-x))
end
#function quadraticneg(x; α=2, β=0.5, γ=0.1)
#    return x + α + β*x + γ*x^2
#end
function quadraticneg(x; shift=3, scaling=0.08)
    return scaling*(x+shift)^2
end
function quadraticpos(x; shift=3, scaling=0.08)
    return scaling*(x-shift)^2 
end
```

```julia
x = -4:0.1:4
print(sigmoid(x[1]), " " , sigmoid(x[end]))
```

```julia
a=5

pl = plot(x, logisticneg.(x), label="", c=:red, linestyle=:dash, linewidth=3)
plot!(x, logisticpos.(x), label=L"$\mathcal{L}_{\mathrm{logistic}}$", c=:red, linewidth=3)

plot!(x, updateneg.(x)/a, label="", c=:orange, linestyle=:dash, linewidth=3)
plot!(x, updatepos.(x)/a, label=L"$\mathcal{L}_{\mathrm{update}}$", c=:orange, linewidth=3)


plot!(x, quadraticneg.(x), label="", c=:blue, linestyle=:dash, linewidth=3)
plot!(x, quadraticpos.(x), label=L"$\mathcal{L}_{\mathrm{logistic'}}$", c=:blue, linewidth=3)


plot!(x, sigmoid.(x), c=:black, label=L"$\sigma$", legend=:top, legendfontsize=15, xtickfontsize=10, ytickfontsize=10, 
xlabel = "logit", xlabelfontsize=15, linewidth=3)

plot!(guidefontsize=20, tickfontsize=15, bottom_margin=5Plots.mm)

save(pl,"lossfunctions")
pl
```
