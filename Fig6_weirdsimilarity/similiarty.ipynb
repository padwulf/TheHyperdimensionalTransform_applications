```julia
include("../plotting.jl")
```

```julia
x = -10:0.1:10
D = 10000
```

```julia
B = randn(D)
b = 2*π*rand(D);
```

```julia
h(x) = cos.(x.*B .+ b).*sin.(x .* B)
hx = reduce(hcat, h.(x));
```

```julia
plot(x, hx[5,:])
plot!(x, hx[7,:])
```

```julia
K = hx'*hx / D
```

```julia
xmarks = [-10, "", -5, "", 0, "", 5, "", 10]
xvalues=range(1,size(K,1), length=length(xmarks))
pl = heatmap(K, c=:greys,colorbar=true, aspect_ratio=:equal,size=(440,400), right_margin = 8Plots.mm)
plot!(xticks=(xvalues, xmarks), yticks=(xvalues, xmarks))
plot!(guidefontsize=20, tickfontsize=15)
save(pl, "heatmap")
pl
```
